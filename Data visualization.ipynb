```python
import pandas as pd
data = pd.read_csv("wine.csv")  #importing a data
```

```python
data.head()
```

```python
data.shape # To show the dimension of th data, the number of columns and rows
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,20)) # the number can also differ
ax = fig.gca()
data.hist(ax=ax)
plt.show  # To make a histogram
```

```python
fig = plt.figure(figsize=(15,20))
ax = fig.gca()
data.plot(ax=ax, kind="density", subplots=True, layout=(4,4), sharex=False)
plt.show  # density curve
```

```python
fig = plt.figure(figsize=(15,20))
ax = fig.gca()
data.plot(ax=ax, kind="line", subplots=True, layout=(4,4), sharex=False)
plt.show  # To make a line plot
```

```python
fig = plt.figure(figsize=(15,20))
ax = fig.gca()
data.plot(ax=ax, kind="bar", subplots=True, layout=(4,4), sharex=False)
plt.show  # Bar plot
```

```python
fig = plt.figure(figsize=(15,20))
ax = fig.gca()
data.plot(ax=ax, kind="box", subplots=True, layout=(4,4), sharex=False)
plt.show  # box plot
```

```python
import seaborn as sb
fig = plt.figure(figsize= (15,20))
ax =fig.gca()
correlations = data.corr(method="pearson")
sb.heatmap(correlations, annot=True, ax=ax)
plt.show  # To get the correlation of the data
```

```python
correlations  # To view the correlation matrix
```

```python
fig = plt.figure(figsize= (15,20))
ax =fig.gca()
correlations = data[["Ash", "Alcohol", "Hue"]].corr(method="pearson")
sb.heatmap(correlations, annot=True, ax=ax)
plt.show  # To slice the correlation of the data, here we want "Ash and Alcohol only"
```

```python
import pandas.plotting as pp 
fig = plt.figure(figsize= (15,15))
ax = fig.gca()
# pp.scatter_matrix(data, ax=ax)  #It gives the scatter plot of the whole data
pp.scatter_matrix(data[["Ash", "Alcohol", "Hue", "Malic acid"]], ax=ax)  # slice to data
plt.show  # To get the scatter plot of the data
```
