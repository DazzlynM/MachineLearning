```python
col0 = iris.data[:,0]
col1 = iris.data[:,1]
col2 = iris.data[:,2]
col3 = iris.data[:,3]
target = iris.target
```

```python
iris_dict = {
    'col0': col0,
    'col1': col1,
    'col2': col2,
    'col3': col3,
    'target': target
}
```

```python
iris_dict
```

```python
import pandas as pd
```

```python
iris_df = pd.DataFrame(data=iris_dict)
```

```python
iris_df
```

```python
target = iris.target_names
```

```python
target
```

```python
iris_df.loc[iris_df['Target']==0, 'Target'] = target[0]
iris_df.loc[iris_df['Target']==1, 'Target'] = target[1]
iris_df.loc[iris_df['Target']==2, 'Target'] = target[2]
```
